## Question 2:
**Now formalize the corresponding optimization version.**


**Input:**

- A complete weighted graph $G = (V, E)$
- A distinguished vertex $s \in V$ representing the warehouse
- A weight function $w : E \rightarrow \mathbb{R}_+$ giving the distance between locations

**Question:**

Find a closed route starting at $s$, visiting every other vertex exactly once, and returning to $s$, such that the total distance is minimized.

**Objective function:**

$$
    \min \sum_{i=0}^{n-1} w(v_i, v_{i+1})
    \quad \text{with } v_n = v_0 = s
$$

where $(v_0, v_1, \dots, v_{n-1})$ is a permutation of the vertices of $V$.

**This is a NP-complete problem : there does not exist a polynomial algorithm to solve it. However, an approximation polynomial algorithm can return a solution that is not far from an optimal solution. Now the following is the detailed description of one such algorithm :**

**2. execute a depth-first traversal on T (you can begin from any vertex of T) such that each edge of T is traversed twice, the result denoted by TD;**

**3. keep in TD only the first occurrence of a vertex and then transform TD into cycle C.**

## Question 5:

**Prove that the solution $s$ returned by the algorithm is a 2-approximation of the optimal solution $s^*$ when the triangle inequality holds: $w(a,c) \leq w(a,b) + w(b,c)$ for any vertices $a, b, c \in V$.**

**Goal:** Show that $\text{total distance}(s) \leq 2 \cdot \text{total distance}(s^*)$.

---

### Step 1: MST weight ≤ optimal tour weight

Let $s^*$ be an optimal TSP tour (Hamiltonian cycle) visiting all vertices and returning to the start. Removing any edge from $s^*$ yields a spanning tree $T'$ (a path that visits each vertex exactly once).

- $\text{weight}(T') = \text{weight}(s^*) - w(e)$ for some edge $e$ with $w(e) \geq 0$
- Since $T'$ is a spanning tree, the MST $T$ has minimum weight among all spanning trees

Therefore:
$$
\text{weight}(\text{MST}) \leq \text{weight}(T') \leq \text{weight}(s^*)
$$

---

### Step 2: DFS traversal doubles the MST

The DFS on the MST traverses each edge exactly twice (once going down, once backtracking). So the total distance of the walk $T_D$ is:
$$
\text{weight}(T_D) = 2 \cdot \text{weight}(\text{MST})
$$

---

### Step 3: Shortcutting does not increase distance (triangle inequality)

The cycle $C$ (our solution $s$) is obtained from $T_D$ by keeping only the **first occurrence** of each vertex and closing the cycle. This amounts to replacing paths like $u \to \cdots \to v$ (where intermediate vertices are skipped) with the direct edge $(u, v)$.

By the triangle inequality:
$$
w(u, v) \leq \text{weight of any path from } u \text{ to } v
$$

So each such shortcut cannot increase the distance. Hence:
$$
\text{total distance}(s) = \text{weight}(C) \leq \text{weight}(T_D)
$$

---

### Conclusion

Combining the three steps:
$$
\text{total distance}(s) \leq \text{weight}(T_D) = 2 \cdot \text{weight}(\text{MST}) \leq 2 \cdot \text{weight}(s^*)
$$

Therefore $\text{total distance}(s) \leq 2 \cdot \text{total distance}(s^*)$, and the algorithm is a **2-approximation**.
